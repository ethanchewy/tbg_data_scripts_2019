```python
# read csv files as df

# Voting record of each county in California
df_ca_poli = pd.read_csv(r'C:\Users\ethan\Desktop\tbg_data_scripts_2019\CA_U.SHouseofRepresentatives.csv', thousands=',')
```

```python
print(df_ca_poli)
```

```python
# GOAL: Process County Level Data to Get Political affiliations of county

# Get Key Values and Set Up New Data Frame
ca_key_vals = df_ca_poli.county.unique()
df_ca_poli_clean = pd.DataFrame(columns=["County", "Republican", "Democrat", "Other"])
df_ca_poli_clean["County"] = ca_key_vals
# print(df_ca_poli_clean)

# Calculate percentage
find_affiliation_tot = lambda df, affil, p_value, v_value : df.loc[df[p_value] == affil][v_value].sum()
for county in ca_key_vals:
    county_df = df_ca_poli.loc[df_ca_poli["county"] == county]
    county_total_votes = county_df["votes"].sum()
    dem_votes = find_affiliation_tot(county_df, "DEM", "party", "votes")
    rep_votes = find_affiliation_tot(county_df, "REP", "party", "votes")

    rep_total = (rep_votes/county_total_votes)*100
    dem_total = (dem_votes/county_total_votes)*100
    df_ca_poli_clean.loc[df_ca_poli_clean["County"] == county, ["Republican"]] = rep_total
    df_ca_poli_clean.loc[df_ca_poli_clean["County"] == county, ["Democrat"]] = dem_total
    df_ca_poli_clean.loc[df_ca_poli_clean["County"] == county, ["Other"]] = 100-(rep_total+dem_total)

df_ca_poli_clean = df_ca_poli_clean[df_ca_poli_clean.County != "Districtwide"]
print(df_ca_poli_clean)

# Remove CountyWide
```
